```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, mean_squared_error, mean_absolute_error, log_loss
```

```python
df = pd.read_csv('C:/Users/BTK/Desktop/Eric/Git/backtest_model/database/STRAT_POLENTOVISK_WIN.csv')
```

```python
df.set_index('date', drop=True, inplace=True)
```

```python
y = df['Dummy'].shift(1).fillna(0)
X = df.drop(['Dummy'], axis=1)
```

```python
X['Tipo'] = np.where(X['Tipo'] == 'buy', 1, 0)
```

```python
scaler = StandardScaler(with_mean=True, with_std=True)
# scaler = MinMaxScaler(feature_range=(-1,1))
```

```python
scaler.fit(X)
```

```python
X_tr = scaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tr, y, test_size=1/3, stratify=y, random_state=123)
```

```python
px.box(X_train)
```

```python
px.box(X_test)
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
```

```python
print(f'Accuracy: {accuracy_score(predict, y_test)}')
print(f'Precision: {precision_score(predict, y_test)}')
print(f'MSE: {mean_squared_error(predict, y_test)}')
print(F'MAE: {mean_absolute_error(predict, y_test)}')
print(F'Log loss: {log_loss(predict, y_test)}')
```

```python
px.violin(abs(predict - y_test), points='all')
```

```python
tam = len(predict)
zeros = np.where(predict - y_test == 0)

print(f'Qtd total de pontos: {tam}')
print(f'Qtd de erros: {tam - len(zeros[0])}')
print(f'Qtd de acertos: {tam - (tam - len(zeros[0]))}')


px.histogram(predict - y_test, marginal='box')
```
